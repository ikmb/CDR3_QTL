## CDR3 and HLA pair-wise matrices for all combinations

#### normalisation of CDR3 freq after excluding germ line AA

```r

###continuation from the cdr3-qtl atricle----
    #we transformed each amino acid frequency into a standard normal distribution (INT-transform).
    #we used this transformed data as in many other eQTL studies.
cdr3_matrix <- function(pdata){ 
    
    normalisation <- pdata %>%
        select(patient_id, AA, relativ_freq) %>%
        group_by(AA) %>%
        mutate(normrate = qnorm((rank(relativ_freq, na.last="keep") - 0.5) / sum(!is.na(relativ_freq)))) %>%
        ungroup() %>%
        select(-relativ_freq) %>% 
        pivot_wider(names_from = AA, values_from = normrate, values_fill = 0)

    return(normalisation)
}


#-------Custom function to align CDR3 AA seq to IMGT------
align_imgt <- function(CDR3){
    AA <- unlist(strsplit(CDR3, ""));
    N_AA_row <- nchar(CDR3) %/% 2 + nchar(CDR3) %% 2;
    N_AA_na <- 18 - nchar(CDR3);
    AAmod <- c(
    AA[1:(N_AA_row)],
    rep(NA,N_AA_na),
    AA[(N_AA_row + 1):nchar(CDR3)]
    )
    
    AAmod_matrix <- matrix(AAmod, ncol=18)
    colnames(AAmod_matrix) <- all_imgt_pos
     
    return(AAmod_matrix)
}

```

### Preparing cdr3-hla-matrices with respect to different sampling size

### Function with MLM, in a full model we exclude the most frequent allele, as the reference allele (as in Ishigaki et al.)

```r

#--------MLM function itself, it takes cdr3-hla matrix, path for the resulting manova df and the replicate default = 1, changes for the power analysis with multiple replicates)----------

mlm <- function(name_pair, cdr3_hla_matrix, pca_hla, dir_results, sample_size = 'all', rep = 1){
    
    amino_acids <-c('A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y') 
    
    cdr3_hla_matrix_pca <- cdr3_hla_matrix %>% 
        mutate(patient_id = as.character(patient_id)) %>%
        inner_join(pca_hla, by = 'patient_id')
    
    without_ref_allele <- cdr3_hla_matrix_pca %>% 
        summarise(across(contains('allele'), sum)) %>% t() %>% as.data.frame() %>% arrange(desc(V1)) %>% slice(-1) %>% rownames()
    aa_in_matrix <- paste0(colnames(cdr3_hla_matrix)[colnames(cdr3_hla_matrix) %in% amino_acids], collapse = ',')
    
    formula_full <- as.formula(paste0(paste0('cbind(',aa_in_matrix,') ~'), paste0(without_ref_allele, collapse = '+'), '+PC1+PC2+PC3'))
    formula_null <- as.formula(paste0('cbind(',aa_in_matrix,') ~', 'PC1+PC2+PC3'))
    
    mod1 <- lm(formula_full, data = cdr3_hla_matrix_pca)
    mod0 <- lm(formula_null, data = cdr3_hla_matrix_pca)
    
    mod1_summary <- summary(mod1)
    mod0_summary <- summary(mod0)
       
    lapply(summary(mod1), function(x){
        capture.output(x, file=paste0('../mmlm_results/mmlm_summary/',name_pair,'_tcelldata_mod1_summary_size_',sample_size,'_rep',rep,'.txt'), append = TRUE)})
    lapply(summary(mod0), function(x){
        capture.output(x, file=paste0('../mmlm_results/mmlm_summary/',name_pair,'_tcelldata_mod0_summary_size_',sample_size,'_rep',rep,'.txt'), append = TRUE)})
    
    manova <- anova(mod1, mod0)

    
    mod1_mvlm <- mvlm(formula_full, cdr3_hla_matrix_pca)
    mod0_mvlm <- mvlm(formula_null, cdr3_hla_matrix_pca)
    
    var_exp_full <- mod1_mvlm$pseudo.rsq["Omnibus Effect",1]
    p_val_full <- mod1_mvlm$pv["Omnibus Effect",1]
    
    var_exp_null <- mod0_mvlm$pseudo.rsq["Omnibus Effect",1]
    p_val_null <- mod0_mvlm$pv["Omnibus Effect",1]
    
    variance_explained <- var_exp_full - var_exp_null
    P_val <-  mapply(c, p_val_full, p_val_null)
    colnames(P_val) <- 'P_val'
    Omnibus = mapply(c, var_exp_full, var_exp_null)
    colnames(Omnibus) <- 'Omnibus'

    manova_df <- data.frame(manova) %>% 
        mutate(pair = name_pair, rep = rep, size = sample_size)
    mvlm_df <- data.frame(Omnibus, model = c('full', 'null'),P_val)
    
    mod1_mvlm_summary <- summary(mod1_mvlm)
    mod0_mvlm_summary <- summary(mod0_mvlm)
    
    
    lapply(summary(mod1_mvlm), function(x){
        capture.output(x,file=paste0('../mmlm_results/mvlm_summary/',name_pair,'_tcelldata_mod1_summary_size_',sample_size,'_rep',rep,'.txt'), append = TRUE)})
    lapply(summary(mod0_mvlm), function(x){
        capture.output(x,file=paste0('../mmlm_results/mvlm_summary/',name_pair,'_tcelldata_mod0_summary_size_',sample_size,'_rep',rep,'.txt'), append = TRUE)})
    
    
    
    #write_tsv(mvlm_df, paste0(dir_results, name_pair,'_sample_size_',sample_size,'_replicate_',replicate,'_omnibus.tsv'))

    manova_df <- bind_cols(manova_df, mvlm_df)
    write_tsv(manova_df, paste0(dir_results,name_pair,'_sample_size_',sample_size,'_replicate_',rep,'.tsv'))
    
    return(manova_df)
    
    
            
        }
```

## Power analysis

### Calling mlm amd mvl functions

```r
read.csv('../mmlm_results/A_105_12_P104_sample_size_100_replicate_1.tsv', sep = '\t')[0,]
```

```r
patients_ids <- unlist(read.table('../patients_ids_wo_duplicates.txt', header = FALSE)$V1)
n_ids <- length(patients_ids)
pca_hla <- read.csv('../pca_hla.tsv', sep = '\t') %>% mutate(patient_id = as.character(patient_id))
dir_results <- '../mmlm_results/'
dir_hla_path <- '../hla_matrices/'
hla_matrices_files <- list.files(dir_hla_path)

manova_df_all_pairs <- read.csv('../mmlm_results/A_105_12_P104_sample_size_100_replicate_1.tsv', sep = '\t')[0,]

for (n in c(600,800,n_ids)){
    for (rep in c(1,2,3)){
        sampling_ids <- sample(x = unlist(patients_ids), size  = n)
        pivoted_cdr3_public_splitted <- pivoted_cdr3_public %>% 
            filter(patient_id %in% sampling_ids) %>%
            group_by(length, IMGT) %>%
            group_split()
        pair_path <- paste0('../cdr3_hla_pairs/',n,'/')
        
        for (L_P_group in seq(1,length(pivoted_cdr3_public_splitted_L_P))){
        
        pivoted_cdr3_L_P <- pivoted_cdr3_public_splitted_L_P[[L_P_group]]
        L <- unique(pivoted_cdr3_public_splitted_L_P[[L_P_group]]$length)
        P <- unique(pivoted_cdr3_public_splitted_L_P[[L_P_group]]$IMGT)
        
        for (hla_file in hla_matrices_files){
            hla <- read.csv(paste0(dir_hla_path,hla_file), sep='\t')
            patients_in_hla_matrix <- unique(pull(hla, patient_id))
            
            pivoted_cdr3_matrix <- cdr3_matrix(pivoted_cdr3_L_P %>% filter(patient_id %in% patients_in_hla_matrix))
            pivoted_cdr3_matrix$L_P <- paste0(L,':',P)
            hla_name <- strsplit(hla_file,'matrix')[[1]][1] 
            name_pair <- paste0(hla_name,L,'_',P)
            cdr3_hla_matrix <- hla %>%
                inner_join(pivoted_cdr3_matrix, by='patient_id')
            write_tsv(cdr3_hla_matrix, paste0(pair_path,name_pair,'_matrix_size_',n,'_rep',rep,'.tsv'))
            
            manova_results <- mlm(name_pair, cdr3_hla_matrix, pca_hla, dir_results, sample_size = n, rep = rep)
            manova_df <- manova_results %>% mutate(pair = name_pair, size = n, rep = rep)
            manova_df_all_pairs <- bind_rows(manova_df_all_pairs, manova_df)

        }
        }
    }
}
```

## Annotation of genes and visualisation part

```r
manova_df_all <- read.csv('../data/TRB_samples/trb_manova_results.tsv', sep = '\t') 
manova_df_all <- manova_df_all %>% filter(!is.na(Pr..F.)) %>% arrange(Pr..F.)
n_tests <- length(cdr3_hla_matrices)
p <- manova_df_all$Pr..F.
bonf <- 0.05/n_tests/20

fdr <- p.adjust(p, method = "fdr", n = length(p)*20)
significant <- manova_df_all %>% filter(Pr..F. <= bonf) %>% arrange(Pr..F.)
```

```r
hla_annotation <- read.csv('../data/TRB_samples/hla_annotations.csv', sep = ';')
hla_annotation <- hla_annotation %>% mutate(start = ifelse(gene %in% c('A','B','C'), start - 24, start - 29), end = ifelse(gene %in% c('A','B','C'), end - 24, end - 29))
```

```r
drb1_annotated_dict <- list(Beta_domain = c(seq(13, 87)), Ig_like_C1_type = c(seq(97, 185)), Antigen_binding = c(57,61,81,82,93), 
                    Glycosylation = 19, Disulfid_bonds	= c(15,79,117,173), Phosphoserine = 237, Crosslinking = 225)
```

```r
drb1 <- manova_df_all %>% filter(gene == 'DRB1') %>% select('Pr..F.','gene','site','length','position') %>% 
    mutate(significant = ifelse((Pr..F. <= bonf), 'yes', 'no'))
drb1_ref <- as.data.frame(do.call(cbind, drb1_annotated_dict)) %>% 
    pivot_longer(names_to = 'feature', values_to = 'site', cols = 1:7) %>% 
    distinct()
drb1_annotated <- left_join(drb1, drb1_ref, by = 'site') %>%
    mutate(interest = ifelse(site %in% c(37,86), 'yes', 'no'), 
           significant = ifelse((Pr..F. <= bonf), 'yes', 'no'))
drb1_site37_site86 <- drb1 %>% filter(site == 37 | site == 86, significant == 'yes') %>% select(-significant)
```
