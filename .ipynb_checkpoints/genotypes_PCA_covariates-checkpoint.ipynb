```r
rsq_long <- rsq %>% pivot_longer(cols=1:(ncol(rsq)-1), names_to = 'allele_2', values_to = 'rsq_value') %>% distinct() 
```

```r
rsq_long %>% filter(rsq_value > 0.5 & rsq_value < 1)
```

```r
hla_data <- fread('../../data/hla/hla_genotypes_matrix.csv')
```

```r

hla_data_labels <- hla_data %>%
    dplyr::select(patient_id) %>% 
    mutate(dataset = case_when(
    grepl("^H", patient_id) ~ "healthy",   
    grepl("^I", patient_id) ~ "IBD")) %>% 
    relocate(dataset, 1)

hla_patients_id <- hla_data$patient_id
hla_matrix <- hla_data %>%
    dplyr::select_if(is.numeric) %>%
    t() %>%
    as.matrix()
colnames(hla_matrix) <- hla_patients_id
pca <- prcomp(hla_matrix, scale. = TRUE)

explained_variance <- pca$sdev^2 / sum(pca$sdev^2)
explained_variance_percent <- explained_variance * 100  # Convert to percentage

pca_df <- data.frame(
  PC = paste0("PC", 1:length(explained_variance_percent)),
  Variance_Explained = explained_variance_percent
)

pca_hla <- as.data.frame(pca$rotation) %>% 
    dplyr::select(c(PC1, PC2, PC3, PC4, PC5, PC6, PC7, PC8, PC9 )) %>% 
    mutate(patient_id = rownames(pca$rotation), 
          dataset = case_when(
              grepl("^H", patient_id) ~ "healthy",    
              grepl("^I", patient_id) ~ "IBD")) %>% 
    relocate(patient_id, 1) %>%
    relocate(dataset, 1)
rownames(pca_hla) <- NULL

pca_plot_all_hla <- ggplot(pca_hla, aes(x = PC1, y = PC2, color = dataset)) +
    geom_point(alpha = 0.5) +
    scale_color_manual(values = c('healthy' = "#673AB7", 'IBD' = "#FF6F61" )) +
    theme_cowplot() + 
    labs(x = paste0('PC1 (',round(pca_df$Variance_Explained[1],2), '%)'), y = paste0('PC2 (',round(pca_df$Variance_Explained[2],2), '%)'),
        title = 'PCA with all HLA alleles') +
      theme(
    plot.title = element_text(size = 12),         # Adjust title size
    axis.title = element_text(size = 10),         # Adjust axis title size
    axis.text = element_text(size = 8),           # Adjust axis label size
    legend.title = element_text(size = 10),       # Adjust legend title size
    legend.text = element_text(size = 8)          # Adjust legend text size
  )
```

```r
pca_plot_all_hla
```

```r
fwrite(pca_hla %>% dplyr::select(-dataset), '../../data/pca_hla_9PCs.tsv', sep = '\t')
```

```r
hla_data <- fread('../data/hla_genotypes_matrix_wo_correlations.csv')
hla_data_labels <- hla_data %>%
    dplyr::select(patient_id) %>% 
    mutate(dataset = case_when(
    grepl("^H", patient_id) ~ "healthy",   
    grepl("^I", patient_id) ~ "IBD")) %>% 
    relocate(dataset, 1)

allele_name <- 'A*02:01'
hla_data_labels$carriers <- hla_data %>% dplyr::select(any_of(allele_name))

hla_patients_id <- hla_data$patient_id
hla_matrix <- hla_data %>%
    dplyr::select_if(is.numeric) %>%
    t() %>%
    as.matrix()
colnames(hla_matrix) <- hla_patients_id
pca <- prcomp(hla_matrix, scale. = TRUE)

explained_variance <- pca$sdev^2 / sum(pca$sdev^2)
explained_variance_percent <- explained_variance * 100  # Convert to percentage

pca_df <- data.frame(
  PC = paste0("PC", 1:length(explained_variance_percent)),
  Variance_Explained = explained_variance_percent
)

pca_hla <- as.data.frame(pca$rotation) %>% 
    dplyr::select(c(PC1, PC2, PC3, PC4, PC5, PC6, PC7, PC8, PC9 )) %>% 
    mutate(patient_id = rownames(pca$rotation), 
          dataset = case_when(
              grepl("^H", patient_id) ~ "healthy",    
              grepl("^I", patient_id) ~ "IBD")) %>% 
    relocate(patient_id, 1) %>%
    relocate(dataset, 1)
rownames(pca_hla) <- NULL
pca_hla <- merge(pca_hla,hla_data_labels, by = c('patient_id', 'dataset'))

pca_plot_wo_corr <- ggplot(pca_hla, aes(x = PC1, y = PC2, color = dataset) ) +
    geom_point(alpha = 0.5) +
    scale_color_manual(values = c('healthy' = "#673AB7", 'IBD' = "#FF6F61" )) + 
    theme_cowplot() +
    labs(x = paste0('PC1 (',round(pca_df$Variance_Explained[1],2), '%)'), y = paste0('PC2 (',round(pca_df$Variance_Explained[2],2), '%)'), 
        title = 'PCA removing correlated HLA alleles') +
      theme(
    plot.title = element_text(size = 12),         # Adjust title size
    axis.title = element_text(size = 10),         # Adjust axis title size
    axis.text = element_text(size = 8),           # Adjust axis label size
    legend.title = element_text(size = 10),       # Adjust legend title size
    legend.text = element_text(size = 8)          # Adjust legend text size
  )

pca_plot_a2 <- ggplot(pca_hla, aes(x = PC1, y = PC2, color = factor(carriers))) +
    geom_point(alpha = 0.5) +
    scale_color_manual(values = c('0' = "black", '1' = "#0F52BA", '2' = "#D81B60")) +
    theme_cowplot() +
    labs(x = paste0('PC1 (',round(pca_df$Variance_Explained[1],2), '%)'), y = paste0('PC2 (',round(pca_df$Variance_Explained[2],2), '%)'), 
        title = 'PCA removing correlated HLA alleles', color = 'Number of A*02:01 alleles') +
      theme(
    plot.title = element_text(size = 12),         # Adjust title size
    axis.title = element_text(size = 10),         # Adjust axis title size
    axis.text = element_text(size = 8),           # Adjust axis label size
    legend.title = element_text(size = 10),       # Adjust legend title size
    legend.text = element_text(size = 8)          # Adjust legend text size
  )
```

```r
pca_both <- pca_plot_all_hla / pca_plot_wo_corr / pca_plot_a2 
pca_both
ggsave('../plots/pca_both_comparing_corr_effect.png', dpi = 600)
```

### Explore the effect of common alleles

```r
freq_alleles <- hla_data[,-1] %>% t() %>% as.data.frame() %>% mutate(row_sum = rowSums(across(where(is.numeric)))) %>% arrange(desc(row_sum)) %>% rownames()
```

```r
freq_alleles
```

```r
allele_name <- 'A*02:01'
hla_data_labels$carriers <- hla_data %>% dplyr::select(any_of(allele))
hla_patients_id <- hla_data$patient_id
hla_matrix <- hla_data %>%
    dplyr::select_if(is.numeric) %>%
    t() %>%
    as.matrix()
colnames(hla_matrix) <- hla_patients_id
pca <- prcomp(hla_matrix, scale. = TRUE)
pca_hla <- as.data.frame(pca$rotation) %>% 
    dplyr::select(c(PC1, PC2, PC3, PC4, PC5, PC6, PC7, PC8, PC9 )) %>% 
    mutate(patient_id = rownames(pca$rotation))
rownames(pca_hla) <- NULL
pca_hla <- merge(pca_hla,hla_data_labels, by = 'patient_id')
pca_plot_a2 <- ggplot(pca_hla, aes(x = PC1, y = PC2, color = carriers)) +
    geom_point() +
    theme_cowplot() +
    labs(titel = allele_name) 

pca_plot_a2
```

```r
for (allele in freq_alleles[1:10]){
    allele_name <- paste0(allele)
    hla_data_labels$carriers <- hla_data %>% dplyr::select(any_of(allele))
    hla_patients_id <- hla_data$patient_id
    hla_matrix <- hla_data %>%
        dplyr::select_if(is.numeric) %>%
        t() %>%
        as.matrix()
    colnames(hla_matrix) <- hla_patients_id
    pca <- prcomp(hla_matrix, scale. = TRUE)
    pca_hla <- as.data.frame(pca$rotation) %>% 
        dplyr::select(c(PC1, PC2, PC3, PC4, PC5, PC6, PC7, PC8, PC9 )) %>% 
        mutate(patient_id = rownames(pca$rotation))
    rownames(pca_hla) <- NULL
    pca_hla <- merge(pca_hla,hla_data_labels, by = 'patient_id')
    pca_plot <- ggplot(pca_hla, aes(x = PC1, y = PC2, color = carriers)) +
        geom_point() +
        theme_cowplot() +
        labs(titel = allele_name)
    
    ggsave(paste0('../plots/PCA/',allele_name,'.jpg'), pca_plot)
    
}
```
